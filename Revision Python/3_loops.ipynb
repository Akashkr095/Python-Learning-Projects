# For Loops

```python
print(1)
print(2)
print(3)
print(4)
print(5)
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,11):
    print("5 x", i,"=", 5*i)
```

# While Loops

```python
i = 1
while i < 6: # condition is checked before each iteration
    print (i)
    i = i + 1
```
